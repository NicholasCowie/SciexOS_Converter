```python
import pandas as pd
```

```python
data = pd.read_excel("Metabolomics_13C_SS_test.xlsx", sheet_name="SciexOS_export")
relevant_columns = ["Plate \nposition", "Sample \nName", "Component Name", "Dilution \nFactor", "Height", "Area"]
data = data[relevant_columns]
```

```python
data["compound"] = data["Component Name"].apply(lambda x: x.split(".")[0])
data["transition"] = data.apply(lambda x: x["Component Name"].split(".")[1][(len(x["compound"])+1):], axis=1)
```

```python
data
```

```python
pivot_table = pd.pivot_table(data, values=["Height", "Area"], index=["Plate \nposition", "Sample \nName", "Dilution \nFactor"], columns=["compound", "transition"])
pivot_table
```
