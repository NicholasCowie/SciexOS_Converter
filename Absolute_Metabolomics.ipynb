```python
import pandas as pd
```

```python
data = pd.read_excel("Quant_Metabolomics.xlsx", sheet_name="SciexOS_export")
relevant_columns = ["Plate Position", "Sample Name", "Component Name", "Dilution Factor", "Height", "Area", "Height Ratio", "Component Group Name", "Calculated Concentration"]
data = data[relevant_columns]
```

```python
data["Component Group Name"]
data["Type"] = data.apply(lambda x: x["Component Name"].split(".")[1], axis=1)
```

```python
data_light = data[data["Type"]=="Light"].reset_index(drop=True)
data_light = data_light.rename(columns={"Height": "Height Light"})
data_heavy = data[data["Type"]=="Heavy"].reset_index(drop=True)
data_heavy = data_heavy.rename(columns={"Height": "Height Heavy"})
data = pd.merge(data_light, data_heavy[["Sample Name", "Component Group Name", "Height Heavy"]], on=["Sample Name", "Component Group Name"])
```

```python
pivot_table = pd.pivot_table(data, values=["Height Heavy", "Height Light", "Height Ratio", "Calculated Concentration"], index=["Plate Position", "Sample Name", "Dilution Factor"], columns=["Component Group Name"])
pt = pivot_table.stack(0).unstack()
```
